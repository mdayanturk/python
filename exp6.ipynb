```python
#Question 1: Write a Python function to find the maximum and minimum numbers from a sequence of numbers

def find_min_max(numbers):
    if not numbers:
        return None, None

    min_val = numbers[0]
    max_val = numbers[0]

    for num in numbers:
        if num < min_val:
            min_val = num
        if num > max_val:
            max_val = num
    return max_val, min_val


sequence1 = [3, 1, 4, 1, 5, 9, 2, 6]
max_num, min_num = find_min_max(sequence1)
print(f"Sequence: {sequence1}")
print(f"Maximum number: {max_num}")
print(f"Minimum number: {min_num}")
```

```python
#Question 2: Write a Python function that takes a positive integer and returns the sum of the cube of all the positive integers smaller than the specified number.

def sum_cubes_smaller(n):
    if not isinstance(n, int) or n <= 0:
        raise ValueError("Input must be a positive integer.")

    total_sum = 0
    for i in range(1, n):
        total_sum += i**3
    return total_sum

# Examples
print(f"Sum of cubes smaller than 4: {sum_cubes_smaller(4)}")
print(f"Sum of cubes smaller than 1: {sum_cubes_smaller(1)}")
print(f"Sum of cubes smaller than 5: {sum_cubes_smaller(5)}")
```

```python
#Question 3: Write a Python function to print $1$ to $n$ using recursion.
def print_1_to_n(n):
    if n <= 0:
        return
    else:
        print_1_to_n(n - 1)
        print(n)

# Examples
print("Printing 1 to 5:")
print_1_to_n(5)
```

```python
#Question 4: Write a recursive function to print the Fibonacci series up to $n$ terms.
def fibonacci_recursive(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    else:
        series = fibonacci_recursive(n - 1)
        series.append(series[-1] + series[-2])
        return series

def print_fibonacci_series(n):
    if n <= 0:
        print("Please enter a positive integer.")
    else:
        fib_series = fibonacci_recursive(n)
        print(f"Fibonacci series up to {n} terms: {fib_series}")

# Example
print_fibonacci_series(7)
```

```python
#Question 5: Write a lambda function to find the volume of a cone.
import math

volume_of_cone = lambda r, h: (1/3) * math.pi * (r**2) * h


radius1, height1 = 3, 7
print(f"Volume of cone with radius {radius1} and height {height1}: {volume_of_cone(radius1, height1):.2f}")
```

```python
#Question 6: Write a lambda function which gives a tuple of max and min from a list
get_max_min_tuple = lambda input_list: (max(input_list), min(input_list))

sample_input = [10, 6, 8, 90, 12, 56]
print(f"Sample input: {sample_input}")
print(f"Sample output: {get_max_min_tuple(sample_input)}")
```

```python
#Question 6: Write a lambda function which gives a tuple of max and min from a list
# a. Keyword Arguments
def greet_person(name, message):
    print(f"Hello, {name}! {message}")

greet_person(name="Alice", message="How are you?")

# b. Default Arguments
def greet_with_default(name, greeting="Hello"):
    print(f"{greeting}, {name}!")

greet_with_default("Charlie")

# c. Variable Length (*args and **kwargs)
def sum_all_numbers(*args):
    total = sum(args)
    print(f"Sum of {args}: {total}")

def display_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

sum_all_numbers(1, 2, 3)
display_info(name="Eve", age=25, city="New York")
```

```python
#Question 8: Write a program to check whether all the values in a dictionary are the same or not using a lambda function.
all_values_same = lambda d: len(set(d.values())) <= 1

dict1 = {'a': 10, 'b': 10, 'c': 10}
dict2 = {'x': 5, 'y': 10, 'z': 5}

print(f"Dictionary 1 - All values same: {all_values_same(dict1)}")
print(f"Dictionary 2 - All values same: {all_values_same(dict2)}")
```

**Q2.**  Create a tuple to store n numberic values and find averages of all types.

```python
n = int(input("Enter number of elements: "))

values = ()
for i in range(n):
    num = int(input("Enter value: "))
    values = values + (num,)

print("Tuple:", values)

if n > 0:
    average = sum(values) / n
    print("Average of values:", average)
else:
    print("Average cannot be calculated (no elements).")
```
